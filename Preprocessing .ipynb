```python
# Input and output file paths (replace with your file paths)
input_file = 'input.csv'
output_file = 'cleaned.csv'

# Function to clean text by removing or replacing non-ASCII characters
def clean_text(text):
    return ''.join(char if ord(char) < 128 else ' ' for char in text)

# Read the CSV file
df = pd.read_csv(input_file, encoding='utf-8')

# Clean the text in all columns
for column in df.columns:
    df[column] = df[column].apply(clean_text)

# Save the cleaned DataFrame to a new CSV file
df.to_csv(output_file, index=False, encoding='utf-8')

print(f'CSV file cleaned and saved to {output_file}')
```
