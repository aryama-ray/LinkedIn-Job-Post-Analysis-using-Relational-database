```python
#removing nonascii characters from all variable having strings
def safe_normalize(x):
    if isinstance(x, str):
        return unicodedata.normalize('NFC', x)
    else:
        return str(x)
```

```python
## Removing nonascii from title
data['title'] = data['title'].apply(safe_normalize)
data['title'] = data['title'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))
```
